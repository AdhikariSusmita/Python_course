```python
a = int(input("enter a="))
b = int(input("enter b="))
o = input("enter + - * /")

def cal():
    if o == "+":
        print(a+b)
    elif o == "-":
        print(a-b)
    elif o == "*":
        print(a*b)
    elif o == "/":
        print(a/b)
    else:
        print("Invalid operator")
    
cal()    
```

```python
#function with arguments

def cal(a,b,o):
    if o == "+":
         print(a+b)
    elif o == "-":
        print(a-b)
    elif o == "*":
        print(a*b)
    elif o == "/":
        print(a/b)
    else:
        print("Invalid operator")
    
a = int(input("enter a="))
b = int(input("enter b="))
o = input("enter + - * /")

cal(a,b,o)

    
```

```python
#return multiple value 

def cal():
    l = 3
    b = 2
    h = 5
    a = l*b
    v = a*h
    return a,v

area,vol = cal()
print("the area is =",area)
print("the volume is =",vol)
```

```python
#functions with argument and return type

def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b


def cal(a,b,o):
    if o == "+":
        print(a+b)
    elif o == "-":
        print(a-b)
    elif o == "*":
        print(a*b)
    elif o == "/":
        print(a/b)
    else:
        print("Invalid operator")
        
a = int(input("enter a="))
b = int(input("enter b="))
o = input("enter + - * /")

    
cal(a,b,o)    
```
